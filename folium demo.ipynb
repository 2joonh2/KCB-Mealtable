```python
nearby_res=pd.read_csv('C:\\Users\\2joon\\Research\\NYC 14K restaurants\\nearby restaurants\\1_nearby_res.csv')
nearby_res.head()
```

```python
df=pd.read_csv('C:\\Users\\2joon\\Research\\NYT reviews_processed_revised.csv', index_col=0)
df
```

```python
df['g_lat'][1],df['g_lon'][1]
```

```python
m = folium.Map(location=[nearby_res['lat'].mean(), nearby_res['lon'].mean()], zoom_start=13)
# ì§€ë„ ì„¤ì •; location: ì¤‘ì•™ì (ìœ„ë„/ê²½ë„), zoom_start: ì‹œìž‘ ì¶•ì 

# Layer of Markers
for idx, row in nearby_res.iterrows():
    lat_ = row['lat']
    lon_ = row['lon']

    folium.Marker(location=[lat_, lon_],
                  tooltip=row['name']).add_to(m)

location=[df['g_lat'][1],df['g_lon'][1]]
folium.Marker(location=location,
                  tooltip='Standard Point').add_to(m)

folium.Circle(radius=800, # meters; about 0.5 mile
              location=location,
              color="#ff7800",
              fill_color='#ffff00',
              fill_opacity=0.2).add_to(m)

m # display map
```
