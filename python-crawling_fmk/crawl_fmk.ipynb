```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://www.fmkorea.com/hotdeal')
bs=BeautifulSoup(html, 'html.parser')

# print(bs)

crawled=bs.find("div", class_="fm_best_widget _bd_pc").find_all('li')
# crawled[0:3]

plist=[]
shops=[]
prices=[]
shippings=[]

for elem in crawled:
    plist.append([elem.find('h3').get_text().replace('\t',''), # title
    'https://fmkorea.com'+elem.find('a').attrs['href']]) # url
    
    metas=elem.find_all('span')
    
    is_shop=0
    is_price=0
    is_shipping=0
    
    for meta in metas:
        str_=meta.get_text().replace('\t','')

        if '쇼핑몰:' in str_:
            shops.append(str_.replace('쇼핑몰:','').strip())
            is_shop=1
        if '가격:' in str_:
            prices.append(str_.replace('가격:','').replace('원','').replace(',','').strip())
            is_price=1
        if '배송' in str_:
            shippings.append(str_.replace('배송:','').replace('0원','무료').strip())
            is_shipping=1
            

    if is_price==0: prices.append('')
    if is_shop==0: shops.append('')           
    if is_shipping==0: shippings.append('') 

    
import pandas as pd
df=pd.DataFrame(plist)
df.columns = ['Title', 'URL']
df['Price']=prices
df['Shop']=shops
df['Shipping']=shippings
df
```
